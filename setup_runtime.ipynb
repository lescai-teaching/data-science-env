# Course environment setup (Google Colab)

Run the install cell once per fresh Colab runtime.

This notebook installs the same **system** and **Python** libraries as the provided Dockerfile (as closely as is practical on hosted Colab), while avoiding known Colab dependency conflicts.


```python
%%bash
set -euo pipefail

echo "[1/3] Installing system packages (apt)..."
sudo apt-get update -y
sudo apt-get install -y --no-install-recommends \
  build-essential \
  cm-super \
  dvipng \
  ffmpeg

echo "[2/3] Installing Python packages (pip, Colab-compatible)..."
python -m pip install -U pip

# Keep Colab's pinned pandas (google-colab requires pandas==2.2.2 in many runtimes),
# and keep protobuf < 6 for TensorFlow / grpcio-status / google-ai-*.
# Also keep dill < 0.3.9 to avoid conflicts with datasets==4.0.0.
python -m pip install -U \
  "protobuf>=5.26.1,<6" \
  "dill>=0.3.0,<0.3.9"

# Install the remaining libraries WITHOUT forcing upgrades of Colab-managed core packages.
python -m pip install \
  altair \
  beautifulsoup4 \
  bokeh \
  bottleneck \
  cloudpickle \
  cython \
  dask \
  h5py \
  ipympl \
  ipywidgets \
  matplotlib \
  numba \
  numexpr \
  openpyxl \
  patsy \
  tables \
  scikit-image \
  scikit-learn \
  scipy \
  seaborn \
  sqlalchemy \
  statsmodels \
  sympy \
  xlrd

echo "[3/3] Warming matplotlib font cache (similar to the Dockerfile)..."
MPLBACKEND=Agg python -c "import matplotlib.pyplot"

echo "Done."
```

## Sanity check
Run the next cell to confirm imports and show versions.

```python
import sys
import importlib

pkgs = [
    'altair',
    'bs4',
    'bokeh',
    'bottleneck',
    'cloudpickle',
    'Cython',
    'dask',
    'dill',
    'h5py',
    'ipympl',
    'ipywidgets',
    'matplotlib',
    'numba',
    'numexpr',
    'openpyxl',
    'pandas',
    'patsy',
    'google.protobuf',
    'tables',
    'skimage',
    'sklearn',
    'scipy',
    'seaborn',
    'sqlalchemy',
    'statsmodels',
    'sympy',
    'xlrd'
]

print('Python:', sys.version)
print('---')

for name in pkgs:
    mod = importlib.import_module(name)
    ver = getattr(mod, '__version__', None)
    if ver is None and name == 'google.protobuf':
        from google.protobuf import __version__ as ver
    print(f"{name:15s} {ver}")
```
